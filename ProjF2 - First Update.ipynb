```python
selected_columns = train_df.select_dtypes(include=['float', 'int']).columns[:-2]
```

```python
plt.figure(figsize=(5, 20))
for i, column in enumerate(selected_columns):
    plt.subplot((len(train_df.columns) // 3) + 1, 3, i + 1)
    sns.boxplot(y=train_df[column])
    plt.title(column)
    plt.tight_layout()

plt.show()
```

- Most smokers are between ages 40 and 60.
- For Cholesterol, we see significant number of outliers on the right side, skewing the data to the right.
- There are also some questionable values such as LDL above 200. This may be due to faulty sensor, invalid input, etc. We will discuss what to do later.

```python
# non-smokers
non_smokers = train_df[train_df['smoking'] == 0]

# smokers
smokers = train_df[train_df['smoking'] == 1]
```

```python
# non-smokers boxplots
plt.figure(figsize=(5, 20))
for i, column in enumerate(selected_columns):
    plt.subplot((len(non_smokers.columns) // 3) + 1, 3, i + 1)
    sns.boxplot(y=non_smokers[column])
    plt.title(column)
    plt.tight_layout()

plt.show()
```

```python
# smokers boxplots
plt.figure(figsize=(5, 20))
for i, column in enumerate(selected_columns):
    plt.subplot((len(smokers.columns) // 3) + 1, 3, i + 1)
    sns.boxplot(y=smokers[column])
    plt.title(column)
    plt.tight_layout()

plt.show()
```

### Common outliers
Surprisingly, the outliers for factors such as LDL are existent for both the smoking and non-smoking demographics. Perhaps, there is confounding variable?
